```python
df
x=df[['YearsExperience']]
y=df[['Salary']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pre=reg.predict(x_test)
```
